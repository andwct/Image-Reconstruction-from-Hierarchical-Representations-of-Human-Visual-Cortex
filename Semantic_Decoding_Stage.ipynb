### prepare fmri data

```python
train_image_list = torch.load("./data/images/images/train_image_list.pt")
print(len(train_image_list))
test_image_list = torch.load("./data/images/images/test_image_list.pt")
print(len(test_image_list))
```

```python

train_fmri = torch.load('./extracted_fmri/sub_1_higher_visual_cortex_training.pt')
train_fmri.shape
```

```python
n_epochs=150
batch_size = 50
lr = 0.0002
b1 = 0.5
b2 = 0.999

np.random.seed(0)
fmri = torch.load('./extracted_fmri/sub_1_higher_visual_cortex_training.pt')
# test_fmri = torch.load("./extracted_fmri/sub_1_higher_visual_cortex_testing.pt")
# labels = torch.load("/home/anderson/Reconstructing-Perceptive-Images-from-Brain-Activity-by-Shape-Semantic-GAN/data/images/images/train_image_category_label_list_revised.pt") 
labels = torch.load("train_cluster20_labels.pt")



print(fmri.shape)
print(labels.shape)

total_blocks = fmri.shape[0]
fmri_size = fmri.shape[1]
print('total blocks:'+str(total_blocks))
print('input fmri size:'+str(fmri_size))

test_num = 10

# cuda = True if torch.cuda.is_available() else False
cuda = False

class Decoder(nn.Module):
    def __init__(self, fmri_size):
        super(Decoder, self).__init__()

        self.model = nn.Sequential(
            nn.Linear(fmri_size, 2048),
            nn.Tanh(),
            nn.Linear(2048, 1024),
            nn.Tanh(),
        )

        self.classify = nn.Sequential(
            nn.Linear(1024, 20),
            nn.Sigmoid(),
        )
        
    def forward(self, fmri):
        fmri_flat = fmri.view(fmri.shape[0], -1)
        semantics = self.model(fmri_flat)
        x = self.classify(semantics)
        return x, semantics

# Use binary cross-entropy loss
adversarial_loss = torch.nn.CrossEntropyLoss()
encoder = Decoder(fmri_size=fmri_size)

# if cuda:
#     encoder.cuda()
#     adversarial_loss.cuda()
encoder.cpu()

# Optimizers
optimizer_E = torch.optim.Adam(encoder.parameters(), lr=lr, betas=(b1, b2))
scheduler_E = torch.optim.lr_scheduler.StepLR(optimizer_E, step_size=10, gamma=0.8, last_epoch=-1)

Tensor = torch.cuda.FloatTensor if cuda else torch.FloatTensor

```

```python

train_fmri = fmri[0:5000]
test_fmri = fmri[5000:6000]


fmri = np.concatenate([train_fmri, test_fmri])
print(fmri.shape)


train_labels = labels[0:5000]
test_labels = labels[5000:6000]
# train_labels = train_labels[rand_id]
labels = np.concatenate([train_labels, test_labels])

fmri = torch.from_numpy(fmri)
fmri = fmri.type(Tensor)

train_fmri = torch.from_numpy(train_fmri)
train_labels = torch.from_numpy(train_labels)
train_labels = train_labels.squeeze()
train_fmri = train_fmri.type(Tensor)
train_labels = train_labels.type(Tensor)

test_fmri = torch.from_numpy(test_fmri)
test_labels = torch.from_numpy(test_labels)
test_labels = test_labels.squeeze()
test_fmri = test_fmri.type(Tensor)
test_labels = test_labels.type(Tensor)


print("train_fmri size", train_fmri.shape)
print("test_fmri size", test_fmri.shape)
print("train_labels", train_labels.shape)
print("test_labels", test_labels.shape)
```

## Train Dataset Semantic Decoding Training

```python
acc_vec = []
train_vec = []
loss_vec = []
for epoch in range(n_epochs):
    for i in range(0,120):
            fmri_data = train_fmri[i*batch_size:(i+1) * batch_size]
            labels_data = train_labels[i*batch_size:(i+1) * batch_size]
            labels_data = labels_data.long()
            optimizer_E.zero_grad()
            predict, _ = encoder(fmri_data)
            e_loss = adversarial_loss(predict, labels_data)
            e_loss.backward()
            optimizer_E.step()
            loss_vec.append(e_loss.item())
            print(
                "[Epoch %d/%d] [Batch %d] [loss: %f] "
                % (epoch, n_epochs, i, e_loss.item())
            )

    if epoch % 10 == 0:
            test_fmri_data = test_fmri
            test_label_data = test_labels.cpu().detach().numpy()
            lbs, _ = encoder(test_fmri_data)
            cpu_labels = lbs.cpu().detach().numpy()
            pred = [np.argmax(lb) for lb in cpu_labels]
            num_correct = (pred == test_label_data).sum()
            acc = num_correct / 6000
            print(acc)
            acc_vec.append(acc)


    scheduler_E.step()
    lrd = optimizer_E.param_groups[0]['lr']


plt.figure()
plt.title('Loss of classification task ')
plt.plot(loss_vec)
plt.xlabel('Epoch num')
plt.ylabel('Loss')
plt.show()
```
